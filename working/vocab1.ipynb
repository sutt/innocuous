```python
from llama_cpp import Llama
llm = Llama(
    model_path="../../data/mistral-7b-instruct-v0.2.Q4_K_M.gguf", 
)
```

```python
def token_bytes(llm, i: int) -> bytes:
    try:
        return llm.detokenize([i])
    except Exception:
        return b""

def escape_for_text(b: bytes) -> str:
    # 1) Decode with surrogateescape to keep raw bytes round-trippable in Unicode
    s = b.decode("utf-8", errors="surrogateescape")
    # 2) Escape non-printables/control bytes visibly (e.g., \x00, \n, \t)
    return s.encode("unicode_escape").decode("ascii")

def dump_vocab_escaped(llm: Llama, path: str) -> None:
    with open(path, "w", encoding="utf-8", errors="strict") as f:
        for i in range(llm.n_vocab()):
            b = token_bytes(llm, i)
            f.write(f"{i}\t{escape_for_text(b)}\n")
```

```python
fn = 'vocab_mistral.txt'

dump_vocab_escaped(llm, fn)
```
